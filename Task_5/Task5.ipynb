```python
# If forgoing ROOT, feel free to use this class of LorentzVectors
class LorentzVector():
    def __init__(self,px,py,pz,mass):
        self.px = px
        self.py = py
        self.pz = pz
        self.energy = np.sqrt(self.px*self.px+self.py*self.py+self.pz*self.pz+mass*mass)
    def P(self):
        return np.sqrt(self.px*self.px+self.py*self.py+self.pz*self.pz)
    def Px(self):
        return self.px
    def Py(self):
        return self.py
    def Pz(self):
        return self.pz
    def Energy(self):
        return self.energy
    def Mag2(self):
        return self.energy*self.energy-(self.px*self.px+self.py*self.py+self.pz*self.pz)
    def Mag(self):
        return np.sqrt(self.Mag2())
    def Print(self):
        print(self.px,self.py,self.pz,self.energy)
    def Negate(self):
        self.px = -self.px
        self.py = -self.py
        self.pz = -self.pz
        self.Print()
    
class Vector3():
    def __init__(self,v1,v2,v3):
        self.v1 = v1
        self.v2 = v2
        self.v3 = v3
    def X(self):
        return self.v1
    def Y(self):
        return self.v2
    def Z(self):
        return self.v3
    def Mag2(self):
        return self.v1*self.v1+self.v2*self.v2+self.v3*self.v3
    def Mag(self):
        return np.sqrt(self.Mag2())
    def Print(self):
        print(self.X,self.Y,self.Z)

def Add_4Vec(vec1,vec2):
    return LorentzVector(vec1.Px()+vec2.Px(),vec1.Py()+vec2.Py(),vec1.Pz()+vec2.Pz(),vec1.Energy()+vec2.Energy())

def Sub_4Vec(vec1,vec2):
    # Vec1 - Vec2
    return LorentzVector(vec1.Px()-vec2.Px(),vec1.Py()-vec2.Py(),vec1.Pz()-vec2.Pz(),vec1.Energy()-vec2.Energy())
```
